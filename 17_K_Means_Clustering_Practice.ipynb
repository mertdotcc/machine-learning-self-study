# K Means Clustering

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Generating Artificial Data using Scikit-learn's built-in tools

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8)
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:,0], data[0][:,1])
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1])
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(15,9))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

## If we have said `KMeans(n_clusters=3)`

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data[0])

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(15,9))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

__Notice the difference!__

## If we have said `KMeans(n_clusters=2)`

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data[0])

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(15,9))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```
